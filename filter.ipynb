```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sta = pd.read_csv("CWB_Stations_171226.csv")
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
stal = list(sta.id)
```

```python
data = ['T Min', 'T Max','Temperature','Precp','RH']
sf = ['tmin.csv','tmax.csv','temperature.csv','precp.csv','rh.csv']
```

```python
for d,fi in zip(data,sf):
    for j in range(len(sta)):
        file = glob("data/Yearly/{}/*.txt".format(stal[j]))
        print("Checking the data on station",stal[j])
        for i,f in enumerate(file):
            df_tmp = pd.read_csv(file[i],index_col=0)
            year = file[i].split('/')[3].split('.')[0]
            df_tmp['ObsTime'].apply(lambda x: '0' + str(x) if int(x) <10 else str(x))
            df_tmp['Time'] = year + '-' + df_tmp['ObsTime'].apply(lambda x: '0' + str(x) if int(x) <10 else str(x)).astype(str)
            if i == 0:
                df = df_tmp
            else:
                df = pd.concat([df,df_tmp])
            dff = pd.DataFrame([[float(a) if is_number(a) else None for a in list(df[d])]],columns= list(df.Time),index=[stal[j]])        
            dff['name'] = len(dff) * ''
            dff.loc[stal[j],"name"] = stal[j]
            DF_tmp = dff
        if j == 0:
            DF = DF_tmp
        else:
            DF = pd.concat([DF,DF_tmp]) 
    DF.to_csv("analysis/" + fi)
```
