```python
species = list(pd.read_csv("../species.csv").scientific_name)
for sp in species:
    file = glob("{}/*".format(sp))
    print(sp)
    for i,f in enumerate(file):
        print(f)
        date = f.split('/')[1].split('_')[0]
        yy = date.split('-')[0]
        mm = date.split('-')[1]
        print(date,yy,mm)
        if i == 0:
            df = pd.read_csv(f)
            df['Time'] = [date] * len(df)
            df['Month'] = [mm] * len(df)
            df['Year'] = [yy] * len(df)
        else:
            dft = pd.read_csv(f)
            dft['Time'] = [date] * len(dft) 
            dft['Month'] = [mm] * len(dft)
            dft['Year'] = [yy] * len(dft)
            df = pd.concat([df,dft])
    df.to_csv("../output_csv/combine/{}.csv".format(sp))
#df[df.EI<0]
```
