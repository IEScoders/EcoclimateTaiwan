```python
from glob import glob
import pandas as pd
```

```python
species = list(pd.read_csv("../species.csv").scientific_name)
```

```python
for sp in species:
    file = glob("{}/*".format(sp))
    print(sp)
    for i,f in enumerate(file):
        print(f)
        date = f.split('/')[1].split('_')[1]
        if i == 0:
            df = pd.read_csv(f)
            df['Time'] = [date] * len(df)
        else:
            dft = pd.read_csv(f)
            dft['Time'] = [date] * len(dft) 
            df = pd.concat([df,dft])
    df.to_csv("../output_csv/combine/{}.csv".format(sp))
#df[df.EI<0]
```
